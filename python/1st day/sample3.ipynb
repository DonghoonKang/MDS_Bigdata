```python
import requests, urllib, os
from bs4 import BeautifulSoup as bs

def download_image(img_src, filename):
    res=urllib.request.urlretrieve(img_src,filename)
    return res
fname='cat'
def crawl_image():
    res = requests.get('https://pixabay.com/en/photos/?q='+fname)
    if res.status_code == 200:
        soup = bs(res.text,'html.parser')
        images = soup.select('#content > div > div > div > div > div > a > img')
        print(images)
        if fname not in os.listdir():
            os.makedirs(fname)
            os.chdir(fname)
        for image in images:
            print(image.attrs)
            image_url = image.attrs['src']
            filename = image_url.split('/')[-1]
            if (filename == "blank.gif"):
                continue
            print(image_url, filename)
#             download_image(image_url,filename)
            with open(filename, "wb") as img_obj:
                img_obj.write(res.content)

crawl_image()
```
